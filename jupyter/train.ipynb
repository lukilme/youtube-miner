```python
numeric_features = [
    "view_count_capped",
    "engagement_rate",
    "like_ratio",
    "comment_ratio",
    "year",
    "month"
]

categorical_features = [
    "macro_category",
    "duration_bucket",
    "country",
    "source_label"
]

numeric_features = [c for c in numeric_features if c in df_clean.columns]
categorical_features = [c for c in categorical_features if c in df_clean.columns]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", 
         Pipeline(steps=[
             ("imputer", SimpleImputer(strategy="median")),
             ("scaler", StandardScaler())
         ]),
         numeric_features
        ),
        ("cat",
         Pipeline(steps=[
             ("imputer", SimpleImputer(strategy="most_frequent")),
             ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
         ]),
         categorical_features
        )
    ],
    remainder="drop"
)

X = df_clean[numeric_features + categorical_features]
X_prepared = preprocessor.fit_transform(X)

feature_names = preprocessor.get_feature_names_out()
X_prepared_df = pd.DataFrame(X_prepared, columns=feature_names, index=X.index)
```
